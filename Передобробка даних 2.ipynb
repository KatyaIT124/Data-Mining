```python
import pandas as pd
```

```python
# Завантажуємо створену таблицю основних даних для подальшої обробки
data_air = pd.read_excel('D:/Jupyter notebook/3 курс/Air_town/Data_air_town.xlsx', sheet_name='Основні дані')
data_air
```

```python
# Завантажуємо створену таблицю основних даних для подальшої обробки
data_index = pd.read_excel('D:/Jupyter notebook/3 курс/Air_town/Data_air_town2.xlsx', sheet_name='Індекси забруднення')
data_index
```

```python
# Перегляд структури та основної інформації
data_air.info()
```

```python
# Заміна значень в колонці "у ГДКс.д." на 0.006 для сульфатів
data_air.loc[data_air['Назва домішки'] == 'Сульфати', 'у ГДКс.д.'] = data_air.loc[data_air['Назва домішки'] == 'Сульфати', 'у ГДКс.д.'].fillna(0.006)
```

```python
# Заміна значень в колонці "у ГДКс.д." на 0.005 для сірководню
data_air.loc[data_air['Назва домішки'] == 'Сiрководень', 'у ГДКс.д.'] = data_air.loc[data_air['Назва домішки'] == 'Сiрководень', 'у ГДКс.д.'].fillna(0.005)
```

```python
# Заміна значень в колонці "у ГДКм.р." на 0.2 для сульфатів
data_air.loc[data_air['Назва домішки'] == 'Сульфати', 'у ГДКм.р.'] = data_air.loc[data_air['Назва домішки'] == 'Сульфати', 'у ГДКм.р.'].fillna(0.2)
```

```python
# Перетворюємо стовпець "Дата" у формат дати
data_air['Дата'] = pd.to_datetime(data_air['Дата'])

# Сортуємо дані за стовпцем "Дата" від меншої до більшої
data_air = data_air.sort_values(by='Дата')
data_air
```

```python
# Перетворюємо стовпець "Дата" у формат дати
data_index['Дата'] = pd.to_datetime(data_index['Дата'])

# Сортуємо дані за стовпцем "Дата" від меншої до більшої
data_index = data_index.sort_values(by='Дата')
data_index
```

```python
data_air.info()
data_index.info()
```

```python
# Створення списку з унікальними назвами домішок
unique = data_air['Назва домішки'].unique()
print(unique)
```

```python
# Створення нової таблиці з домішками та класами небезпеки, до яких вони відносяться
data_class = pd.DataFrame({'Назва домішки': ['Завислі речовини', 'Дiоксид сiрки', 'Оксид вуглецю', 'Дiоксид азоту',
                                             'Фтористий водень', 'Амiак', 'Формальдегiд', 'Фенол', 'Сажа', 'Хлористий водень',
                                             'Оксид азоту', 'Сiрководень', 'Сульфати', 'Анілін'],
                           'Клас небезпеки': [3, 3, 4, 3, 2, 4, 2, 2, 3, 2, 3, 2, 4, 2]})
data_class
```

```python
# Зміна типу даних
data_class['Клас небезпеки'] = data_class['Клас небезпеки'].astype('int8')
# Перегляд модифікованої структури та основної інформації
data_class.info()
```

```python
# Запис у excel-файл на різні листи
with pd.ExcelWriter('D:/Jupyter notebook/3 курс/Air_town/Air_town_data3.xlsx') as writer:
    data_air.to_excel(writer, sheet_name='Основні дані', index=False)
    data_index.to_excel(writer, sheet_name='Індекси забруднення', index=False)
    data_class.to_excel(writer, sheet_name='Класи небезпеки', index=False)
```
