```python
# Групування даних за містами
grouped = data_merge_new.groupby('Місто')

# Функція для обчислення кореляції та формування порівняльної таблиці
def compute_correlations(method):
    correlations = {}
    for city, group in grouped:
        # Вибір колонок для обчислення кореляції
        relevant_columns = ['tavg', 'tmin', 'tmax','prcp',	'wdir',	'wspd',	'pres', 'qср, мг/м3', 'qмакс, мг/м3 ', 'Індекс забруднення атмосфери']
        correlation_matrix = group[relevant_columns].corr(method=method)
        correlations[city] = correlation_matrix
    
    comparison_data = []
    for city, corr in correlations.items():
        for col1 in corr.columns:
            for col2 in corr.columns:
                if col1 != col2:
                    comparison_data.append({
                        'Місто': city,
                        'Параметри': f'{col1} - {col2}',
                        f'Кореляція ({method})': corr.loc[col1, col2]
                    })
    return pd.DataFrame(comparison_data)

# Обчислення кореляцій за Пірсоном, Спірменом та Кендаллом
comparison_df_pearson = compute_correlations('pearson')
comparison_df_spearman = compute_correlations('spearman')
comparison_df_kendall = compute_correlations('kendall')

# Об'єднання результатів в одну таблицю
comparison_df2 = comparison_df_pearson.merge(
    comparison_df_spearman, on=['Місто', 'Параметри'], how='outer'
).merge(
    comparison_df_kendall, on=['Місто', 'Параметри'], how='outer'
)

# Виведення порівняльної таблиці
comparison_df2
```

```python
# Функція для виділення найвищих і найнижчих кореляційних значень
def extract_extreme_correlations(corr_matrix):
    # Залишаємо лише нижній трикутник матриці, щоб уникнути дублікатів
    lower_triangle = corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=-1).astype(bool))
    
    # Розгладжуємо нижній трикутник матриці в серії
    lower_triangle = lower_triangle.unstack().dropna()
    
    # Знаходимо найбільші і найменші кореляції
    max_correlation = lower_triangle.max()
    min_correlation = lower_triangle.min()
    
    # Отримуємо пари параметрів з найбільшими і найменшими кореляціями
    max_corr_pair = lower_triangle.idxmax()
    min_corr_pair = lower_triangle.idxmin()
    
    return max_corr_pair, max_correlation, min_corr_pair, min_correlation

# Зберігаємо результати для всіх міст
extreme_correlations = []

for city, corr_matrix in correlations_2.items():
    max_corr_pair, max_correlation, min_corr_pair, min_correlation = extract_extreme_correlations(corr_matrix)
    extreme_correlations.append({
        'Місто': city,
        'Найвища кореляція (пара)': f'{max_corr_pair[0]} - {max_corr_pair[1]}',
        'Найвища кореляція (значення)': max_correlation,
        'Найнижча кореляція (пара)': f'{min_corr_pair[0]} - {min_corr_pair[1]}',
        'Найнижча кореляція (значення)': min_correlation
    })

extreme_correlations_df = pd.DataFrame(extreme_correlations)

# Виводимо результати
extreme_correlations_df
```
