```python
import pandas as pd
```

```python
# Завантаження та перегляд даних
data = pd.read_excel('D:/Jupyter notebook/4 курс/Air_town/Забруднення/air_town_january_21.xlsx', header = 2)
data
```

```python
# Перегляд структури та основної інформації
data.info()
```

```python
# Створення фільтра з метою вилучення рядків, що містять індекси забруднення повітря
filter_condition = data['Місто'].str.startswith('Індекс забруднення атмосфери')& data.isnull().any(axis=1)

# Створення нового датафрейму із видаленими рядками з індексами забруднення
data_new = data[filter_condition].copy()
data_new = data_new.filter(regex=("Місто|Індекс забруднення атмосфери"))
print(data_new)

# Видалення рядків з індексами забруднення з вихідного датафрейму
data.drop(data[filter_condition].index, inplace=True)
print(data)
```

```python
# Видалення порожніх рядків у вихідному датафреймі
data = data.dropna(how='all')
data
```

```python
# Створення списку унікальних значень міст
unique_cities = data['Місто'].unique()
print(unique_cities)
```

```python
# Перейменування стовпця у таблиці з індексами
data_new.rename(columns={"Місто": "Індекс забруднення атмосфери"}, inplace=True)
data_new
```

```python
# Створення нового стовпця з містами, що відповідають індексам
data_new['Місто'] = unique_cities
data_new
```

```python
# Редагування стовпця з індексами, щоб залишились тільки числові значення
data_new['Індекс забруднення атмосфери'] = data_new['Індекс забруднення атмосфери'].replace(
    {'Індекс забруднення атмосфери: ': ''}, regex=True)
data_new
```

```python
# Зміна послідовності колонок
data_new = data_new.reindex(columns=['Місто', 'Індекс забруднення атмосфери']).reset_index(drop=True)
data_new
```

```python
# Додавання колонки з датою
data_new['Дата'] = '31.01.2021'
data_new
```

```python
# Перегляд структури та основної інформації у таблиці з індексами
data_new.info()
```

```python
# Зміна типу даних
data_new['Індекс забруднення атмосфери'] = data_new['Індекс забруднення атмосфери'].str.replace(',', '.').astype('float32')
data_new.info()
```

```python
# Перегляд модифікованої структури
data_new.info()
```
